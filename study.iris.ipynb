```python

X=amostra_treino.drop(columns=['Species'])
X_teste=amostra_teste.drop(columns=['Species'])
# iloc é com índices
# loc é com os nomes das colunas


# mulclasse
# treino
Y_virginica_tr=amostra_treino['Species'].map(lambda x: 1  if x=='virginica' else 0)
Y_versicolor_tr=amostra_treino['Species'].map(lambda x: 1  if x=='versicolor' else 0)
Y_setosa_tr=amostra_treino['Species'].map(lambda x: 1  if x=='setosa' else 0)
#Teste
Y_virginica_ts=amostra_teste['Species'].map(lambda x : 1 if  x=='virginica' else 0)
Y_versicolor_ts=amostra_teste['Species'].map(lambda x: 1 if x=='versicolor' else 0)
Y_setosa_ts=amostra_teste['Species'].map(lambda x: 1 if x=='setosa' else 0)


amostra_treino_mult=pd.concat([Y_virginica_tr.rename("virginica"),Y_versicolor_tr.rename("versicolor"),Y_setosa_tr.rename("setosa"),amostra_treino],axis=1)
amostra_teste_mult=pd.concat([Y_virginica_ts.rename('virgnica'),Y_versicolor_ts.rename('versicolor'),Y_setosa_ts.rename('setosa') ,amostra_teste],axis=1)
```
